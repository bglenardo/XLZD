```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
M_Xe = 131.293  # g/mol
M_N = 931.5 # MeV/nucleon
rho_Xe = 2.942  # g/cm^3
nat_abundance_Xe136 = 0.0892

N_A = 6.022e23  # 1/mol
livetime = 10 # years
```

```python
def calculate_bkgfree_sensitivity(mass_tonne, livetime, n_evts = 2.44):
    mass_g = mass_tonne * 1e6  # Convert tonne to grams
    n_atoms = (mass_g / M_Xe) * N_A * nat_abundance_Xe136  # Number of target nuclei
    
    sensitivity = n_atoms * livetime * np.log(2) / n_evts # 90% CL for 0 observed events
    
    return sensitivity
```

```python
# Plot the sensitivity upper limit vs. mass up to 80 tonnes for a natural xenon detector
livetime = 10.  # years

masses = np.linspace(0., 80, 100)  # Masses from 1 to 80 tonnes
sensitivities = [calculate_bkgfree_sensitivity(mass, livetime) for mass in masses]

plt.rcParams.update({'font.size': 14})

plt.figure(figsize=(10, 6))
plt.plot(masses, sensitivities, label='Background-free sensitivity', color='blue')
plt.yscale('log')
plt.xlabel('Fiducial mass [tonnes]')
plt.ylabel('90% CL half-life sensitivity [yr] \n (assumes 2.44 events)')

plt.ylim(1e26, 1e29)
plt.xlim(0., 80)
plt.grid()
plt.title('Background-free Sensitivity vs. Fiducial Mass \n for natXe Detector with 10yr livetime')
```
