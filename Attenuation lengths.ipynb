```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# IMport Xe photon scattering data
```

```python
# Import xe scattering data using numpy
data = np.loadtxt('Xe photon scattering data.txt', delimiter = '\t',skiprows=1 )

indices = {'Energy (MeV)': 0,
           'Coherent (cm2/g)': 1,
           'Incoherent (cm2/g)': 2,
            'Photoelectric (cm2/g)': 3,
            'Pair production nuclear (cm2/g)': 4,
            'Pair production electron (cm2/g)': 5,
            'Total attenuation w/ coherent (cm2/g)': 6,
            'Total attenuation w/o coherent (cm2/g)': 7
            }
```

```python
plt.rcParams.update({'font.size': 14})
plt.rcParams.update({'figure.figsize': (8,6)})

density = 3. # g/cm3 for LXe at 178K

# Plot attenuation length, total w/ coherent, along with all the subcomponents
plt.plot(data[:,indices['Energy (MeV)']], \
    1./(data[:,indices['Total attenuation w/ coherent (cm2/g)']]*density), \
        label='Total w/ coherent',\
            color='black', linewidth=3)


plt.xscale('log')
plt.yscale('log')

# Plot subcomponents
# plt.plot(data[:,indices['Energy (MeV)']], 1./(data[:,indices['Coherent (cm2/g)']]*density), label='Coherent')
plt.plot(data[:,indices['Energy (MeV)']], 1./(data[:,indices['Incoherent (cm2/g)']]*density), label='Incoherent (Compton)')
plt.plot(data[:,indices['Energy (MeV)']], 1./(data[:,indices['Photoelectric (cm2/g)']]*density), label='Photoelectric')
plt.plot(data[:,indices['Energy (MeV)']], 1./(data[:,indices['Pair production nuclear (cm2/g)']]*density), label='Pair production nuclear')
# plt.plot(data[:,indices['Energy (MeV)']], 1./(data[:,indices['Pair production electron (cm2/g)']]*density), label='Pair production electron')


plt.xlabel('Photon Energy (MeV)')
plt.ylabel('Attenuation Length (cm)')

# Make sure the axis labels are actually like 1, 10, 100 rather than 10^0, 10^1, 10^2
# plt.gca().xaxis.set_major_formatter(plt.ScalarFormatter())
# plt.gca().yaxis.set_major_formatter(plt.ScalarFormatter())
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.legend()

plt.ylim(0.01, 100.)
plt.xlim( 0.01, 30.)

plt.title('Photon Attenuation Lengths in LXe (from NIST XCOM)')

```

```python
# Plot all the curves on the same plot, with attenuation factor (Inverse of attenuation length) on the y-axis, and log-log scale  
plt.plot(data[:,indices['Energy (MeV)']], data[:,indices['Total attenuation w/ coherent (cm2/g)']]*density, label='Total w/ coherent', color='black', linewidth=3)
plt.xscale('log')
plt.yscale('log')


plt.xlabel('Photon Energy (MeV)')
plt.ylabel('Attenuation Factor (cm^2/g)')


plt.grid(which='both', linestyle='-', linewidth=0.5, color=(0.8,0.8,0.8,0.5))

# plt.plot(data[:,indices['Energy (MeV)']], data[:,indices['Coherent (cm2/g)']]*density, label='Coherent')
plt.plot(data[:,indices['Energy (MeV)']], data[:,indices['Incoherent (cm2/g)']]*density, label='Incoherent (Compton)')
plt.plot(data[:,indices['Energy (MeV)']], data[:,indices['Photoelectric (cm2/g)']]*density, label='Photoelectric')
plt.plot(data[:,indices['Energy (MeV)']], data[:,indices['Pair production nuclear (cm2/g)']]*density, label='Pair production (nuclear)')
# plt.plot(data[:,indices['Energy (MeV)']], data[:,indices['Pair production electron (cm2/g)']]*density, label='Pair production electron')

plt.legend(framealpha=1.)
plt.xlim(1e-3,1e3)
plt.ylim(0.003, 1000.)

plt.title('Photon Attenuation Factors in LXe (from NIST XCOM)')

```
