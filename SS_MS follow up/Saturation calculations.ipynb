```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.rcParams['font.size'] = 14

gain = 2e6
C_per_e = 1.6e-19  # C/e
e_per_C = 1 / C_per_e  # e/C

C_base = 10e-9  # 10 nF
V_base = 121.7 # V

stored_charge = C_base * V_base  # C

num_photons = np.linspace(0.,3e6,1000)
fraction_of_stored_charge = (num_photons * gain * C_per_e) / stored_charge

plt.plot(num_photons, fraction_of_stored_charge, '-b')
plt.xlabel('Number of Photons Detected')
plt.ylabel('Fraction of Stored Charge Used')
# plt.yscale('log')
plt.xscale('log')
plt.ylim(0.,1.)
plt.xlim(1e3, 3e6)
plt.grid()
plt.savefig('stored_charge_vs_photons.png', dpi=300, bbox_inches='tight')

```
